```python
n=[3000,2000,1000,500,100]
seq=[[264262877,261644892,261994880,259661238,261886294],[78370507,77458849,77781988,78214283,78163361],[8422373,8532437,8420423,8498529,8455171],[1470346,1489857,1438404,1471358,1978831],[8431,8915,9103,16838,11883]]
prll=[[93337801,97772982,93286262,93334772,93407173],[31387001,31741606,31924745,32244623,32739916],[5077819,5017101,4925438,5418184,4920299],[1083354,1206488,1219718,1152806,1009665],[7528,10959,8106,6214,9197]]
```

```python
av_seq=[]
av_prll=[]
for i in range(len(seq)):
    sum=0
    for j in range(5):
        sum+=seq[i][j]
    av_seq.append(sum/5)
    
for i in range(len(seq)):
    sum=0
    for j in range(5):
        sum+=prll[i][j]
    av_prll.append(sum/5)
```

```python
av_seq
```

```python
av_prll
```

```python
n.sort()
av_seq.sort()
av_prll.sort()
```

```python
print(n, av_seq, av_prll)
print(*av_seq)
```

```python
plt.plot(n,av_seq,label="sequential")
plt.plot(n,av_prll,label="parallel")
plt.xlabel("input size(n)")
plt.ylabel("running time(in ms)")
plt.title("comparison b/w running times for sequential and parallel programs",pad=20)
plt.legend()
plt.tight_layout()
plt.savefig("/home/saksham/Documents/sys_prac/observ.png")
plt.show()
```
